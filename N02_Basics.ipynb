## Lists

```python
int_list = [1, 2, 3, 4, 5]
string_list = ['foo', 'bar', 'baz']
print(int_list)
print(string_list)
```

```python
print(int_list[:2])
print(int_list[2:])
print(int_list[-2])

int_list[0] = 100
print(int_list)

new_int_list = int_list[:2] + int_list[-2:]
print(new_int_list)

# But, be careful:

print(int_list[2] + int_list[-2])

# Instead:

print([int_list[2]] + [int_list[-2]])

# Or:

int_list.append(6)
print(int_list)
```

## Flow Control

```python
a = 0
if (a == 0): 
    print('A is zero')
elif (a == 1):
    print('A is one')
else:
    print('A is neither zero nor one')
```

## Functions

```python
def add(a, b):
    return a+b
add(1, 2)
```

```python
# Variable scopes are important!

c = 0
def add(a, b):
    c = 100
    print('Inside function', c)
    return a + b
add(1, 2) # Note that this is just called and does not print anything!
print('Outside function', c)
```

```python
# Arguments can have default values

def add(a, b=100):
    return a + b
print(add(1, 2))
print(add(1))
```

```python
# Arguments can be anything!

def add(a, b):
    return a + b

def subtract(a, b): # We can re-use arguments in a new scope
    return a - b

def arbitrary(f, a, b):
    return f(a, b)

print(arbitrary(add, 1, 2))
print(arbitrary(subtract, 1, 2))
```
